## Next, we can clean up our movie titles

```python
tmp = movies.title.str.extract('(.*) \(([0-9]+)\)')
```

```python
tmp.apply(lambda x:x[0] if len(x) > 0 else None)
```

```python
tmp.apply(lambda x: x[0][:40] if len(x) > 0 else None)
```

```python
movies['year'] = tmp[1]
```

```python
movies['short_title'] = tmp[0]
```

```python
movies.head()
```

## Now we have created a new column titled "short title" that doesnt include the year. Now, we can look to merge our datasets with the pandas merge function. We can start by inner joining our tables users and ratings on the user id column. This will merge the datasets that have a one to many relationship within this column, and an inner join should merge these datasets on matching values.

```python
full_df = pd.merge(users, ratings, on = "user_id")
```

```python
full_df.head()
```

```python
full_df.shape
```

## The first merge is shown above, as we can see we retain the same amount of observations but at 4 new columns. We can now merge this new dataset with the dataset "movies", but this time on the column "movie_id".

```python
full_df = pd.merge(full_df, movies, on = "movie_id")
```

```python
full_df.head()
```

```python
full_df.shape
```

## As expected, we can merge our datasets to now add 4 new columns to our dataset. Our new dataset, "full_df" includes the columns from all 3 of the datasets we read in, but the same amount of observations as we had ratings with 1,000,209. Now, we can find the highest rated movie, as well as the best movie for date night based on some condition.

```python
ratings = full_df.groupby('movie_id', as_index=False)['rating'].mean()
```

```python
ratings
```

```python
ratings = ratings.sort_values(by="rating", ascending = False)
```

```python
ratings.head(20)
```

## Above we have sorted all of the movies by the mean of all of their ratings, which can be identified by thier movie_id. We can see a handful of movies that have recieved 5 stars, any of which can be found by name, such as the top movie with an id of 926. 

```python
full_df[full_df["movie_id"] == 926]
```

## We can see the top 5 star movie is titled "All About Eve" and we could do this for any of the highest rated movies in our dataset.

## Now we will look at finding a good rated movie for date night based on whatever criteria we so choose. First, we want to find all of the movies that were voted at least 4 stars on average by both men and women. Then we can filter out certain genres that we feel are better suited for a date night.

```python
male_df = full_df[full_df["gender"] == "M"]
```

```python
male_df.head()
```

```python
female_df = full_df[full_df["gender"] == "F"]
```

```python
female_df.head()
```

```python
male_ratings = male_df.groupby('movie_id', as_index=False)['rating'].mean()
```

```python
male_ratings = male_ratings.sort_values(by="rating", ascending = False)
```

```python
male_ratings.head(20)
```

```python
female_ratings = female_df.groupby('movie_id', as_index=False)['rating'].mean()
```

```python
female_ratings = female_ratings.sort_values(by="rating", ascending = False)
```

```python
female_ratings.head(28)
```

## We can see there are 11 movies that males have at a perfect 5 star rating and 27 movies that females have at a perfect 5 star rating. we can slice these series so only the 5 star movies remain from the list of descending rankings.

```python
male_ratings = male_ratings[:11]
```

```python
male_ratings.head(20)
```

```python
female_ratings = female_ratings[:27]
```

```python
female_ratings.head(30)
```

```python
combined_ratings = pd.merge(male_ratings, female_ratings, on = "movie_id")
```

```python
combined_ratings.head()
```

## We can see there is only one movie that has a 5 star rating for both men and women. The title of this movie can be found by locating the movie that has a movie id of 787.

```python
full_df[full_df["movie_id"] == 787]
```
